````python
model = genai.GenerativeModel('gemini-1.5-flash')

prompt = f"scrap information about dog breed from this page,: {html_content}"

response1 = model.generate_content(prompt)
````

````python
print(response1.text[:50])
````

````python
import re
from bs4 import BeautifulSoup

# Assuming response1.text contains the raw HTML content
raw_html = response1.text

# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(raw_html, 'html.parser')

# Extract the relevant content
# For this example, let's assume you want to extract information about dog breeds

# Find all relevant tags, assuming the dog breed information is in <div> tags with a specific class
breed_divs = soup.find_all('div', class_='breed-info')  # Replace 'breed-info' with the actual class name if different

# Extract and print the text content of these tags
breeds = []
for div in breed_divs:
    breed_name = div.find('h2').text  # Assuming breed names are in <h2> tags
    breed_description = div.find('p').text  # Assuming breed descriptions are in <p> tags
    breeds.append({'name': breed_name, 'description': breed_description})

# Display the extracted breeds
for breed in breeds:
    print(f"Breed Name: {breed['name']}\nDescription: {breed['description']}\n")
````
